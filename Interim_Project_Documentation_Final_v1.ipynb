# **Methodology Approach**

The methodology for this project adopted a structured and technology-driven approach, leveraging a robust **ETL (Extract, Transform, Load)** pipeline and multiple advanced tools to manage, analyze, and visualize the datasets effectively.

image.png











**Data Extraction** was achieved programmatically using the **Kaggle API** and the **kagglehub** library, which streamlined the retrieval of two datasets: one in **CSV** format (Mobile Device Usage and User Behavior Dataset) and another in **JSON** format (Phones 2024 Dataset). The JSON dataset, offering detailed device specifications, was processed using Python’s **pandas.json_normalize()** to flatten nested data structures into a tabular format suitable for relational databases.

**Data Transformation** was conducted extensively using Python’s **Pandas** library. The cleaning process involved removing irrelevant columns, standardizing column names, and addressing inconsistencies, such as missing or null values, by populating averages or excluding incomplete rows. JSON and CSV datasets were merged using an inner join on the device_model attribute to create a unified dataset. Explicit typecasting was implemented to standardize data types after merging, ensuring numerical columns like display_size retained their correct float64 data type. This phase also involved the preparation of the dataset into three logical tables: **mobile_device**, focusing on device specifications; **mobile_user**, capturing user demographics and behavior classes; and **usage_detail**, highlighting metrics like app usage time, screen-on time, and data consumption.

**Data Loading** involved storing the transformed data into PostgreSQL databases, managed using the **sqlalchemy** Python library for database connectivity. Two PostgreSQL instances were deployed:

1. A local instance named "**UnCalledFour**" used for testing and development.
2. A cloud instance on **Aiven for PostgreSQL** named "**uncalledfour**", providing scalability and remote access.

**Database design** emphasized efficient storage and query execution by incorporating relational schema principles. **Primary** keys were designated (e.g., device_model), and **foreign** keys maintained referential integrity between the tables. **Constraints** were applied to ensure data quality, while indexing improved query performance.

image.png






The above Entity-Relationship (ER) Diagram illustrates the detailed schema with three interconnected tables: **mobile_device**, **mobile_user**, and **usage_detail**. The mobile_device table contains information about various mobile device models, including attributes like device_model (primary key), operating_system, weight, and display_size. Each device can be associated with multiple users through the mobile_user table, establishing a **one-to-many (1:M)** relationship. The mobile_user table stores user details such as user_id (primary key), age, gender, user_behaviour_class, and the **device_model** they own, which serves as a **foreign key** linking back to the mobile_device table.

The mobile_user table is further connected to the usage_detail table in a **one-to-one (1:1)** relationship. **Each user has exactly one record in the usage_detail table, which provides insights into their device usage patterns**. Key attributes in the usage_detail table include user_id (primary key and foreign key), app_usage_time, screen_on_time, battery_drain, no_apps_installed, and data_usage. The foreign key relationships ensure consistency across the tables, enforcing that user and device details align properly with the associated records.

This schema design enforces data integrity through **primary and foreign key constraints**, ensuring unique identifiers for devices and users while maintaining consistent references between tables. **Not Null constraints** on fields like device_model and user_id ensure essential data is always recorded.

**Visualization** of the findings was conducted using **Matplotlib**, a Python library for creating detailed graphs and plots. Insights were presented through visualizations such as bar charts, scatter plots, and distribution graphs, focusing on key metrics like app usage time, screen-on time, and battery drain. These visualizations highlighted patterns and correlations, providing actionable insights into the factors impacting battery performance.

## Data Extraction Using Kagglehub Library

```python
# Download Mobile device usage and user behavior dataset

import kagglehub
import pandas as pd
import os

# Dataset URL: # https://www.kaggle.com/datasets/valakhorasani/mobile-device-usage-and-user-behavior-dataset
dataset_name = "valakhorasani/mobile-device-usage-and-user-behavior-dataset"

try:
    # Download dataset
    path = kagglehub.dataset_download(dataset_name)
    files = os.listdir(path)
    print("Downloaded:", files)

    # If no CSV
    csv_files = [file for file in files if file.endswith('.csv')]
    if not csv_files:
        raise FileNotFoundError("No CSV file found in the dataset!")

    # Df
    csv_file_path = os.path.join(path, csv_files[0])
    df = pd.read_csv(csv_file_path)
    print(df.head())

except Exception as e:
    print("An error occurred:", str(e))
```

## Data Transformation

```python
# This section of code deals with the ETL Process - T (Transformation)
# Data cleaning, Standardizing Formats, Data Type Conversion, Data Integration

import json
import pandas as pd

# Load the JSON file
with open('./data/data.json', 'r') as file:
    data = json.load(file)

# Normalize the JSON data to flatten nested JSON objects into a DataFrame using json_normalize() method of Pandas library
df = pd.json_normalize(data)

# Extracted only relevant phone models (Apple iPhone 12,OnePlus 9, Samsung Galaxy S21 5G)
filtered_df = df.loc[(df['phone_model']=='Apple iPhone 12') | (df['phone_model']=='OnePlus 9') | (df['phone_model']=='Samsung Galaxy S21 5G'),['phone_model', 'specs.Body.Weight','specs.Display.Size']]
print("Extracted dataframe: \n")
print(filtered_df)

# Making a copy of the filtered df
filtered_df2 = filtered_df.copy()

print("\nTransformation STARTED...")

# Standardization of Naming Convention
filtered_df2['phone_model'] = filtered_df2['phone_model'].replace('Apple iPhone 12', 'iPhone 12')
filtered_df2['phone_model'] = filtered_df2['phone_model'].replace('Samsung Galaxy S21 5G', 'Samsung Galaxy S21')

# Cleaning the Weight Column - Data Type Conversion
filtered_df2.loc[:, 'weight'] = filtered_df2['specs.Body.Weight'].str.extract(r'^(.*?)\s+g')[0]
filtered_df2['weight'] = filtered_df2['weight'].astype(int)

# Cleaning the Display Size Column - Data Type Conversion & Column Name Rename
filtered_df2.loc[:, 'display_size'] = filtered_df2['specs.Display.Size'].str.extract(r'^(.*?)\s+inches')[0]
filtered_df2['display_size'] = filtered_df2['display_size'].astype(float)

# Removing unwanted columns
df_json =filtered_df2.drop(columns=['specs.Body.Weight','specs.Display.Size'])

# Renaming the Column Name
df_json = df_json.rename(columns={df_json.columns[0]: 'device_model'})

print("Transformation COMPLETED !!\n")
print("Extracted dataframe after transformation: \n")
print(df_json)

print("\n Populating missing values...\n")

# Populate missing values - Google Pixel 5 & Xiaomi Mi 11
new_data = pd.DataFrame({'device_model': ['Google Pixel 5', 'Xiaomi Mi 11'], 'weight': [151, 196], 'display_size': ['6.0', '6.81']})
df_json = pd.concat([df_json, new_data], ignore_index=True)
print(df_json)
```

```python
# Dataframes Merge - device_details_df (dataframe from .csv) & df_json
mobile_device_df = pd.merge(device_details_df, df_json, on='device_model', how='inner')
print(mobile_device_df)
```

```python
# Local  Postgres - Define Primary Keys, Foreign Keys

import psycopg2

try:
    # Connect to the database
    connection = psycopg2.connect(
        dbname="UnCalledFour",
        user="postgres",
        password="admin",
        host="localhost",
        port="5432"
    )
    cursor = connection.cursor()

    # Add a primary key for mobile_device Table
    cursor.execute("""
        ALTER TABLE mobile_device
        ADD CONSTRAINT pk_device_model PRIMARY KEY (device_model);
    """)

    # Add a primary key for mobile_user Table
    cursor.execute("""
        ALTER TABLE mobile_user
        ADD CONSTRAINT pk_user_id PRIMARY KEY (user_id);
    """)

    # Add a primary key for usage_detail Table
    cursor.execute("""
        ALTER TABLE usage_detail
        ADD CONSTRAINT pk_user_id2 PRIMARY KEY (user_id);
    """)

    # Add the 1 to 1 relationship mobile_user and usage_detail
    cursor.execute("""
        ALTER TABLE usage_detail
        ADD CONSTRAINT fk_usage_detail_id
        FOREIGN KEY (user_id)
        REFERENCES mobile_user (user_id)
        ON DELETE CASCADE;
    """)

    # Add foreign key on mobile_user Table
    cursor.execute("""
        ALTER TABLE mobile_user
        ADD CONSTRAINT fk_device_model
        FOREIGN KEY (device_model)
        REFERENCES mobile_device(device_model);
    """)

    # Commit the changes
    connection.commit()
    print("Primary & Foreign keys applied successfully!")

except Exception as e:
    print("An error occurred:", e)

finally:
    # Close connection
    if cursor:
        cursor.close()
    if connection:
        connection.close()
```

```python
#Plot 4.1: Avg Battery Drain By OS and User Behavior Class (Grouped Bar Chart)
#-------------------------------------------------------------------------------------

query = """
select operating_system,user_behavior_class, avg(battery_drain) AS battery_drain from
(select * from usage_detail left join mobile_user using(user_id) left join mobile_device using(device_model))
group by operating_system, user_behavior_class
order by avg(battery_drain)
"""

#df_temp = pd.read_sql(query, db_engine)    # Local Instance
df_temp = pd.read_sql(query, postgres_uri)  # Aiven


# Pivot data for easier plotting
pivot_data = df_temp.pivot(index="user_behavior_class", columns="operating_system", values="battery_drain")

# Plot a grouped bar chart
ax = pivot_data.plot(kind="bar", figsize=(10, 6))

# Add labels to each bar
for container in ax.containers:
    # Attach labels to each bar
     ax.bar_label(container, fmt='%.f', label_type='edge')

plt.title("Average Battery Drain by Operating System and User Behavior Class")
plt.xlabel("User Behavior Class")
plt.ylabel("Average Battery Drain (mAh/day)")
plt.legend(title="Operating System")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
